```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(df.drop('Private',axis=1))
```

```python
model.cluster_centers_
```

```python
# creating a new column called cluster which is 1 for private school and
# 0 for public school
```

```python
def converter(private):
    if private=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster']=df['Private'].apply(converter)
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(df['Cluster'],model.labels_))
print(confusion_matrix(df['Cluster'],model.labels_))
```

```python
# Not so bad considering the algorithm is purely using the features 
# to cluster the universities into 2 distinct groups!
# Hopefully you can begin to see
# how K Means is useful for clustering un-labeled data!
```
